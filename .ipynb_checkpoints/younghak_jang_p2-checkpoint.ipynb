ANSWER:

**a. Why doesn't nearest neighbors work well for this problem?**

k-nearest neighbor in sklearn uses Minkowski distance metric, which is an n-dimensional version of Euclidean distance, as the default distance metric. So applying this metric into vectors from CountVectorizer makes serious problem because,
* if a certain word appear in a document once and many times in another they are considered far away from each other, while
* if there exists a document that has only one word that happen to be the next word in the vocabulary, they are considered close to each other even if the next word doesn't have any relationship with the previous word.

For example, let's say there are only two words in the vocabulary, [word, letter].
* Doc 1: 'word'
* Doc 2: 'word word word word'
* Doc 3: 'letter'

Minkowski distance metric will say that Doc 3 is closer to Doc1 than Doc2, but intuitively Doc1 and Doc2 have higher probability of having the same label than Doc1 and Doc3. So k-nearest neighbor shouldn't work as well as other algorithms that puts no meaning in adjacency of words in the vocabulary.

**b. Any ideas why logistic regression doesn't work as well as Naive Bayes?**

**c. Briefly explain the relationship between the sum and the value of C**

Low C ( < 0.01) makes most of the weights 0 and as C increase the sum of weights squared increases.

###(5) Try to improve the logistic regression classifier by passing a custom preprocessor to CountVectorizer. 
The preprocessing function runs on the raw text, before it is split into words by the tokenizer. Your preprocessor should try to normalize the input in various ways to improve generalization. For example, try lowercasing everything, replacing sequences of numbers with a single token, removing various other non-letter characters, and shortening long words. If you're not already familiar with regular expressions for manipulating strings, see https://docs.python.org/2/library/re.html, and re.sub() in particular. With your new preprocessor, how much did you reduce the size of the dictionary?

For reference, I was able to improve dev F1 by 2 points.

[4 pts]

```python
def empty_preprocessor(s):
    return s

def better_preprocessor(s):
### STUDENT START ###

    ### this function removes stopwords and punctuation marks, and lower cases everything
    
    # first process the stopwords
    # copied English stopwords list from http://www.ranks.nl/stopwords/ and saved it into a text file
    text_file = open("stopword.txt", "r")
    stopword = text_file.read().split('\t')
    
    # we want to replace only full words, so add a space in front and end of the stopwords
    # i.e. since 'at' is a stopword, if we don't do this it will make 'threat' into 'thre'
    for i in range(0, len(stopword)) :
        stopword[i] = " " + stopword[i] + " "
    
    # initiate a list of processed messages for output
    pruned_s = []
    
    for i in range(0, len(s)) :
        processed = s[i]
        # replace the stopwords with space
        for word in stopword :
            processed = processed.lower().replace(word.lower(), " ")
            
        # lower case everything and remove punctuation marks
        pruned_s.append(re.sub("[.,-_=()!?@#$%^&*\"`\'{}<>]", " " , processed.lower()))
    
    return pruned_s

### STUDENT END ###


def P5():
### STUDENT START ###
    # preprocess the train and dev set
    pruned_train_data = better_preprocessor(train_data)
    pruned_dev_data = better_preprocessor(dev_data)
    
    # apply Count Vectorizer
    cv = CountVectorizer()
    train_pruned = cv.fit_transform(pruned_train_data)
    pruned_vocab = cv.vocabulary_
    
    cv2 = CountVectorizer(vocabulary = pruned_vocab)
    dev_pruned = cv2.fit_transform(pruned_dev_data)
    
    # run logistic regression with C = 0.5
    lr = LogisticRegression(C=0.5)
    lr.fit(train_pruned, train_labels)
    dev_lr = lr.predict(dev_pruned)
    
    # calculate the f-score
    f1_pruned = metrics.f1_score(dev_labels, dev_lr)
    
    
    ### for reference, run logistic regression with C = 0.5 using raw train/dev data
    cv3 = CountVectorizer()
    train_vocab = cv3.fit_transform(train_data)
    default_vocab = cv3.vocabulary_
    cv4 = CountVectorizer(vocabulary = default_vocab)
    dev_vocab = cv4.fit_transform(dev_data)
    
    lr2 = LogisticRegression(C=0.5)
    lr2.fit(train_vocab, train_labels)
    dev_lr2 = lr2.predict(dev_vocab)
    
    f1_default = metrics.f1_score(dev_labels, dev_lr2)
    
    # print comparison
    print "Preprocessing Result: "
    print "f-score improved from %.4f to %.4f (%.4f improvements)" % (f1_default, f1_pruned, f1_pruned-f1_default)
    print "vocabulary size decreased from %d to %d" % (train_vocab.shape[1], train_pruned.shape[1])
    
### STUDENT END ###
P5()
```

###(6) 
The idea of regularization is to avoid learning very large weights (which are likely to fit the training data, but not generalize well) by adding a penalty to the total size of the learned weights. That is, logistic regression seeks the set of weights that minimizes errors in the training data AND has a small size. The default regularization, L2, computes this size as the sum of the squared weights (see P3, above). L1 regularization computes this size as the sum of the absolute values of the weights. The result is that whereas L2 regularization makes all the weights relatively small, L1 regularization drives lots of the weights to 0, effectively removing unimportant features.

Train a logistic regression model using a "l1" penalty. Output the number of learned weights that are not equal to zero. How does this compare to the number of non-zero weights you get with "l2"? Now, reduce the size of the vocabulary by keeping only those features that have at least one non-zero weight and retrain a model using "l2".

Make a plot showing accuracy of the re-trained model vs. the vocabulary size you get when pruning unused features by adjusting the C parameter.

Note: The gradient descent code that trains the logistic regression model sometimes has trouble converging with extreme settings of the C parameter. Relax the convergence criteria by setting tol=.01 (the default is .0001).

[4 pts]

```python
def P6():
    # Keep this random seed here to make comparison easier.
    np.random.seed(0)

    ### STUDENT START ###
    # initialize Count Vectorizer
    cv = CountVectorizer()
    train_vocab = cv.fit_transform(train_data)
    feat_string = cv.get_feature_names()
    
    ### Train a logistic regression model using a "l1" penalty
    lr_l1 = LogisticRegression(penalty='l1', tol=.01)
    lr_l1.fit(train_vocab, train_labels)
    l1_coef = lr_l1.coef_
    
    # Output the number of learned weights that are not equal to zero. 
    l1_count = 0
    for i in range(0, l1_coef.shape[0]) :
        for j in range(0, l1_coef.shape[1]) :
            if l1_coef[i][j] != 0 : l1_count += 1
    print "Number of learned weights that are not equal to zero(L1 regularization): %d, %.1f%% of all weights" % (l1_count, float(l1_count)/(l1_coef.shape[0] * l1_coef.shape[1]) * 100)

    ### Train a logistic regression model using a "l2" penalty
    lr_l2 = LogisticRegression(penalty='l2', tol=.01)
    lr_l2.fit(train_vocab, train_labels)
    l2_coef = lr_l2.coef_
    
    # Output the number of learned weights that are not equal to zero. 
    l2_count = 0
    for i in range(0, l2_coef.shape[0]) :
        for j in range(0, l2_coef.shape[1]) :
            if l2_coef[i][j] != 0 : l2_count += 1
    print "Number of learned weights that are not equal to zero(L2 regularization): %d, %.1f%% of all weights" % (l2_count, float(l2_count)/(l2_coef.shape[0] * l2_coef.shape[1]) * 100)
    
    ### ANSWER: while weights for L2 regularization are all(107516) non-zero, L1 regularization makes only 1815(1.7% of 107516) weights non-zero
    
    ### Reduce the size of the vocabulary by keeping only those features that have at least one non-zero weight and retrain a model using "l2"
    # prepare vocabulary with non-zero weights in L1 regularization
    non_zero_feat = []
    for i in range(0, l1_coef.shape[1]) :
        if (l1_coef[0][i] != 0) or (l1_coef[1][i] != 0) or (l1_coef[2][i] != 0) or (l1_coef[3][i] != 0) :
            non_zero_feat.append(feat_string[i])

    cv_pruned = CountVectorizer(vocabulary = non_zero_feat)
    train_pruned = cv_pruned.fit_transform(train_data)

    # train logistic regression using L2 regularization with vocabulary from L1 regularization
    lr_l2_pruned = LogisticRegression(penalty='l2', tol=.01)
    lr_l2_pruned.fit(train_pruned, train_labels)
    
    # using the same vocabulary predict dev data
    cv_pruned_dev = CountVectorizer(vocabulary = non_zero_feat)
    dev_vocab = cv_pruned_dev.fit_transform(dev_data)
    dev_lr_pruned = lr_l2_pruned.predict(dev_vocab)
    
    # calculate accuracy
    correct = 0
    for i in range(0, len(dev_lr_pruned)) :
        if dev_lr_pruned[i] == dev_labels[i] : correct += 1
    
    print "Accuracy for L2 regularization with vocabulary from non-zero weight features of L1 regularization: %.4f" % (float(correct)/len(dev_lr_pruned))
    
    
    ### Iterate above process (train/predict using L2 regularization with vocabulary from L1 regularization) while varying C
    # keep track of number of vocabulary and accuracy while varying C
    c_value = [0.001, 0.01, 0.1, 0.5, 1.0, 2.0, 10.0]   # c=0.0001 makes all weights 0
    c_vocab = []
    c_acc = []
    
    # iterate with varying C
    for c in c_value :
        # train using L1 regularization and get vocabulary
        lr_l1_c = LogisticRegression(penalty='l1', C=c, tol=.01)
        lr_l1_c.fit(train_vocab, train_labels)
        l1_c_coef = lr_l1_c.coef_
        
        non_zero_feat = []
        for i in range(0, l1_c_coef.shape[1]) :
            if (l1_c_coef[0][i] != 0) or (l1_c_coef[1][i] != 0) or (l1_c_coef[2][i] != 0) or (l1_c_coef[3][i] != 0) :
                non_zero_feat.append(feat_string[i])
    
        # train and fit using L2 regularization
        cv_c_pruned = CountVectorizer(vocabulary = non_zero_feat)
        train_c_pruned = cv_c_pruned.fit_transform(train_data)
    
        lr_l2_c_pruned = LogisticRegression(penalty='l2', tol=.01)
        lr_l2_c_pruned.fit(train_c_pruned, train_labels)
    
        # predict on dev data and calculate accuracy
        cv_c_pruned_dev = CountVectorizer(vocabulary = non_zero_feat)
        dev_c_vocab = cv_c_pruned_dev.fit_transform(dev_data)
        dev_lr_c_pruned = lr_l2_c_pruned.predict(dev_c_vocab)
    
        correct = 0
        for i in range(0, len(dev_lr_c_pruned)) :
            if dev_lr_c_pruned[i] == dev_labels[i] : correct += 1
        
        # keep track of vocabulary size and accuracy for plotting
        c_vocab.append(len(non_zero_feat))
        c_acc.append(float(correct)/len(dev_lr_c_pruned))
        print "Vocabulary size and accuracy for C = %.3f: %d, %.4f" % (c, len(non_zero_feat), float(correct)/len(dev_lr_c_pruned))

    # plot vocabulary size and accuracy
    plt.plot(c_vocab, c_acc)
    plt.show()
    
    ### STUDENT END ###
P6()
```

Make predictions on the dev data and show the top 3 documents where the ratio R is largest, where R is:

maximum predicted probability / predicted probability of the correct label

What kinds of mistakes is the model making? Suggest a way to address one particular issue that you see.

```python
# function for sorting nested list by value
def getKey(item) :
    return item[1]
```

```python
def P7():
    ### STUDENT START ###
    
    # Term Frequency - Inverse Document Frequency considers term frequency and how many documents have that term at the same time.
    # So even if a term appears many times in a document, if that term appears in many other documents at the same time it's tf-idf value is relatively low
    # Vice versa, if a term doesn't appear many times in a document but there are only few documents with that word it's tf-idf value is relatively high
    # Count Vectorizer only considers term frequency and doesn't care in how many documents that term appears
    
    # initialize Tfidf Vectorizer
    tfidf = TfidfVectorizer()
    train_tfidf = tfidf.fit_transform(train_data)
    tfidf_dev = TfidfVectorizer(vocabulary = tfidf.vocabulary_)
    dev_tfidf = tfidf_dev.fit_transform(dev_data)
    feat_string = tfidf.get_feature_names()
    
    # train and fit logistic regression with C = 100
    lr = LogisticRegression(C=100)
    lr.fit(train_tfidf, train_labels)
    dev_pred = lr.predict(dev_tfidf)
    dev_pred_proba = lr.predict_proba(dev_tfidf)

    # calculate accuracy
    correct = 0
    for i in range(0, len(dev_pred)) :
        if dev_pred[i] == dev_labels[i] : correct += 1
        
    print "Accuracy: %.4f" % (float(correct)/len(dev_pred))
    
    # calculate the R ratio for each data where R is maximum predicted probability / predicted probability of the correct label
    R = []
    for i in range(0, dev_pred_proba.shape[0]) :
        correct_label = dev_labels[i]
        correct_pred_proba = dev_pred_proba[i][correct_label]   # predicted probability of the correct label
        max_pred_proba = max(dev_pred_proba[i][0], dev_pred_proba[i][1], dev_pred_proba[i][2], dev_pred_proba[i][3])   # maximum predicted probability
        R.append([i, max_pred_proba / correct_pred_proba])
    
    # sort by R value
    R_sorted = sorted(R, key = getKey, reverse = True)
    
    # get messages with top 3 R value
    for i in range(0, 3) :
        print '*' * 50
        print 'Dev Data #%d, R ratio: %.4f\n' % (R_sorted[i][0], R_sorted[i][1])
        print dev_data[R_sorted[i][0]]
        print '\nPredicted Label: ', newsgroups_train.target_names[dev_pred[R_sorted[i][0]]]
        print 'Correct Label: ', newsgroups_train.target_names[dev_labels[R_sorted[i][0]]]
    
    
    coef_array = lr.coef_
    max5_index = np.ndarray(shape=(4,10))
    
    # print top 5 weights from each label for investigation
    print "\nfeature weight table"
    for i in range(0,4) :
        max5_index[i] = np.argsort(-coef_array[i])[:10]
        print "Features with largest weight for", newsgroups_train.target_names[i]
        for j in range(0,10) : 
            print "%-15s %8.4f %8.4f %8.4f %8.4f" %(feat_string[int(max5_index[i][j])], 
                                               coef_array[0][max5_index[i][j]], coef_array[1][max5_index[i][j]], 
                                               coef_array[2][max5_index[i][j]], coef_array[3][max5_index[i][j]])
    ### STUDENT END ###
P7()
```

ANSWER: 

Since tf-idf accounts for the number of documents certain word is in, if a word appears in only few documents and those documents happen to have the same label, then the weights for that word could be exaggerated even if its meaning doesn't have anything to do with the label. 

On the weights table, I suspect *bobby(alt.atheism), enviroleague(alt.atheism), hi(comp.graphics), 42(comp.graphics), dc(sci.space), nick(sci.space)* are some examples of such cases.

One way to fix this could be setting the min_df to prevent exaggerating such terms.
